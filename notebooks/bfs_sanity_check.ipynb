### 0.1 Adjacency Matrix

```python
#Generate an adjacency matrix given a list of vertices and edges (undirected)
#Time complexity to traverse in this form: O(n^2)

def adjMatrix(graph, n):
    M = []
    #Create a loop to fill an nxn matrix with zeroes
    for i in range(n):
        M.append([0]*n)
    #Create a loop that adds a 1 for each connected edge
    for u,v in graph:
        #Replace element at u,v with 1, creating a symmetric matrix
        M[u][v] = 1
        M[v][u] = 1
    return M

#Test the code above and print the adjacency matrix
n = 7
graph = [[0,1],[0,2],[0,3],[1,2],[2,3],[2,4],[3,4],[4,5],[4,6]]
adjMatrix(graph,n)
```

### 0.2 Adjacency List

```python
#Generate an adjacency list given a list of vertices and edges (undirected)
#Time complexity to traverse graph in this form: O(n) (O(V+E))

#Get defaultdict from collections, which assigns an empty list as a value to a missing key (no errors thrown)
from collections import defaultdict

def adjList(graph):
    #Initialize a dictionary to represent the adjacency list
    D = defaultdict(list)
    for u, v in graph:
        #Since hashmap/dict cannot have duplicates, we can map multiple values to a key without duplicate entries.
        D[u].append(v)
        D[v].append(u)
    return D

#Test the code above and print the adjacency list
graph = [[0,1],[0,2],[0,3],[1,2],[2,3],[2,4],[3,4],[4,5],[4,6]]
adjList(graph)    
```

### 0.3 Breadth-First Search (BFS)

```python
#Implement a simple BFS by passing a graph in adjacency list form and specifying start node. Return a valid bfs.

def BFS(graph, start):
    
    #Make adjacency list from graph
    D = adjList(graph)
    print(D)
    #import deque from collections. Double-ended queues allow enqueuing and dequeuing in O(1) time.
    from collections import deque
    #Initialize a queue which will track visited nodes
    q = deque()
    #Store a set with valid BFS. This ensures checking with O(1) time complexity.
    visited = set()
    #Create a list to store order of visitation.
    bfs = []
    #Add the start node to the set and the queue
    visited.add(start)
    q.append(start)
   
    #The queue tracks nodes that need to be explored. Once a node has been visited, it is popped from the queue
    # and explored. Once all nodes are visited and explored, the queue becomes empty and the BFS is ended.
    while len(q) != 0:
        explore = q.popleft()
        bfs.append(explore)
        print(bfs)
        #Once a node has been popped from the queue, it will be explored. The adjacency list stores all 
        # adjacent nodes for a single node.
        adj = D[explore]
        #For loop to explore the current node
        for i in adj:
            #Check whether i already exists in the BFS set. If not, add to set and the queue as the next node to visit.
                if i not in visited:
                     visited.add(i)
                     q.append(i)
    return bfs

#Test the code above and print a valid BFS
graph = [[0,1],[0,2],[0,3],[1,2],[2,3],[2,4],[3,4],[4,5],[4,6]]
BFS(graph,4)
```

### 0.4 BFS With Target Node

```python
#Modify BFS such that the search is stopped once a target node is reached.
#Inputs: graph (list), start node, end node
#Output(s): Valid BFS (list)

def targetBFS(graph,start,target):

    #Make adjacency list from graph
    D = adjList(graph)
    print(D)
    #import deque from collections. Double-ended queues allow enqueuing and dequeuing in O(1) time.
    from collections import deque
    #Initialize a queue which will track visited nodes
    q = deque()
    #Store a set with valid BFS. This ensures checking with O(1) time complexity.
    visited = set()
    #Create a list to store order of visitation.
    order = []
    #Add the start node to the set and the queue
    visited.add(start)
    q.append(start)
   
    #The queue tracks nodes that need to be explored. Once a node has been visited, it is popped from the queue
    # and explored. Once all nodes are visited and explored, the queue becomes empty and the BFS is ended.
    while q:
        
        explore = q.popleft()
        order.append(explore)
        print(order)
        
        #Check whether the node to be explored is the target. If yes, end the search and return the BFS
        if explore == target:
             return order
        else:
         
         #Once a node has been popped from the queue, it will be explored. The adjacency list stores all adjacent nodes for a single node.
         adj = D[explore]
         #For loop to explore the current node
         for i in adj:
             #Check whether i already exists in the BFS set. If not, add to set and the queue as the next node to visit.
                if i not in visited:
                    visited.add(i)
                    q.append(i)
    return order

#Test the code above and print a valid BFS which ends at the target
graph = [[0,1],[0,2],[0,3],[1,2],[2,3],[2,4],[3,4],[4,5],[4,6]]
targetBFS(graph,5,6)
```

### 0.5 BFS with Path Reconstruction

```python
#Modify the BFS algorithm such that the nodes leading to the target can be stored as it is explored. 
# This can later be reconstructed to form the path to the target node from the start.

#Inputs: graph (list), start node, end node
#Output(s): Valid BFS (list)

def pathBFS(graph,start,target):

    #Make adjacency list from graph
    D = adjList(graph)
    print(D)
    
    #import deque from collections. Double-ended queues allow enqueuing and dequeuing in O(1) time.
    from collections import deque

    #Initialize a queue which will track visited nodes
    q = deque()
    
    #Store a set with valid BFS. This ensures checking with O(1) time complexity.
    visited = set()
    
    #The start node is a special case and doesn't have a predecessor, so we assign it a value of None in the hashmap.
    parent = {start:None}
    #Add the start node to the set and the queue
    visited.add(start)
    q.append(start)
   
    #The queue tracks nodes that need to be explored. Once a node has been visited, it is popped from the queue
    # and explored. Once all nodes are visited and explored, the queue becomes empty and the BFS is ended.
    while q:
        
        explore = q.popleft()
    
        #Once a node has been popped from the queue, it will be explored. The adjacency list stores all adjacent nodes for a single node.
        adj = D[explore]
        #For loop to explore the current node
        for i in adj:
            #Check whether i already exists in the BFS set. If not, add to set and the queue as the next node to visit.
            if i not in visited:
                visited.add(i)
                q.append(i)
                #Add the node to the dictionary and track its parent
                parent[i] = explore
                if i == target:
                    #Path Reconstruction Code

                    #Create a path vector to store the path (in reverse, from target -> start)
                    path = []
                    path.append(target)
                    j = 0
                    while parent[path[j]] is not None:
                        path.append(parent[path[j]])
                        j = j + 1
                    path.reverse()
                    return {"path" : path ,
                                "length" : len(path) - 1}

    #If the target node doesn't exist, return None
    return None

#Test the code above and print a valid BFS which ends at the target
graph = [[0,1],[0,2],[0,3],[1,2],[2,3],[2,4],[3,4],[4,5],[4,6]]
pathBFS(graph,5,6)
```
