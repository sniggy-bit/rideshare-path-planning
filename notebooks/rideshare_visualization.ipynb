### Visualization of Search Algorithms (for Debugging and comparison)

```python
#Create a visualization tool of the search using matplotlib to show the search process in real-time. 
# This can help you understand how the algorithm explores the search space and how it converges to a solution. 
import matplotlib.pyplot as plt
import numpy as np
import src.graph.grid as Grid
from src.graph.bfs import bfs_shortest_path
from src.routing.route_evaluator import path_cost_finder
from tests.test_static_routes import test_route_evaluator
from src.routing.static_planner import is_valid_sequence, route_generator
from src.routing.events import RideRequest, RequestSet

def visualize_search(grid, start, goal):
    # Create a figure and axis for plotting
    fig, ax = plt.subplots()

    #Create a grid for testing
    test_grid = Grid(3,3)
    
    # Get the shortest path using BFS
    path = bfs_shortest_path(grid, start, goal)
    
    # Plot the grid
    ax.imshow(grid, cmap='gray')
    
    # Plot the start and goal points
    ax.scatter(start[1], start[0], c='green', marker='o', label='Start')
    ax.scatter(goal[1], goal[0], c='red', marker='x', label='Goal')
    
    # Plot the path if it exists
    if path:
        path_x = [point[1] for point in path]
        path_y = [point[0] for point in path]
        ax.plot(path_x, path_y, c='blue', label='Path')
    
    # Set labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('Search Visualization')
    ax.legend()
    
    # Show the plot
    plt.show()


```
